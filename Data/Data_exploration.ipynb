```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
crew = pd.DataFrame(columns=['credit_id', 'department', 'gender', 'id', 'job', 'name', 'movie_id'])
cast = pd.DataFrame(columns=['cast_id', 'character', 'credit_id', 'gender','id', 'name', 'order'])
for index, row in credits.iterrows():
    a = pd.read_json(row['cast'])
    a['movie_id'] = row['movie_id']
    c = pd.read_json(row['crew'])
    c['movie_id'] = row['movie_id']
    cast = cast.append(a)
    crew = crew.append(c)
```

```python
crew.head()
```

```python
crew.reset_index(inplace=True, drop=True)
```

```python
crew.head()
```

```python
def f(x):
    r = '['
    for i, s in enumerate(x):
        if i > 0:
            r = r + ','
        r = r + str(int(s))
    return r + ']'
```

```python
crew2 = crew.groupby(['department', 'gender', 'id', 'job', 'name'])
```

```python
crew2 = crew2.agg({'movie_id':f})
```

```python
crew2.reset_index(inplace=True)
```

```python
crew2.head()
```

```python
crew2.to_csv('crew.csv')
```
