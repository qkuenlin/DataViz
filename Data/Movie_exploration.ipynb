```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
genres = pd.DataFrame(columns=['id', 'name'])
for s in movies.genres:
    t = pd.read_json(s)
    genres = genres.append(t)
```

```python
unique_genres = genres.drop_duplicates().set_index("id")
```

```python
unique_genres
```

```python
unique_genres.to_csv("genre.csv")
```

```python
movies.genres = movies.genres.apply(lambda x: pd.read_json(x)["id"].tolist() if "id" in x else [])
```

```python
keywords = pd.DataFrame(columns=['id', 'name'])
for s in movies.keywords:
    t = pd.read_json(s)
    keywords = keywords.append(t)
```
